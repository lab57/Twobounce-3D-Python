```python
def parser():
    lines = 0
    vts = []
    fs = []
    with open("./CUBE_TST.obj", "r") as f:
        lines = f.readlines()
    for line in lines:
        line = line.strip()
        if line[0:2]=="vt": #texture coordinate
            res = line.split(" ")
            vts.append([float(n) for n in res[1:]])

        if line[0] == "f":
            spl = line.split(" ")[1:]

            #extract texutre coordinates
            tis = [index-1 for index in [float(group.split("/")[1]) for group in spl]]
            face = [vts[int(i)] for i in tis]
            fs.append(face)
    #print(fs)
    return fs
parser()
```
